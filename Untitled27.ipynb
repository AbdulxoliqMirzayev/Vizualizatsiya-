```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/uzbekistan.csv", index_col=0)
df.columns=['maydon','aholi']
df
```

```python
plt.figure(figsize=(16,6))
plt.title("O'zbekiston Respublikasi tarkibidagi hududlar soni")
plt.xlabel("Hududlar")
plt.ylabel("Aholi soni")
plt.xticks(rotation=90)
plt.grid()
sns.barplot(x=df.index, y=df['aholi'])
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0,parse_dates=[0],usecols=[0,1])
df.head()
```

```python
df['month']=[i.month for i in df.index]
df.head()
```

```python
df['year']= [i.year for i in df.index]
df.head()
```

```python
df2 = df.groupby(['month','year']).max()
df2.head()
```

```python
df2.unstack(level=1)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data=df2.unstack(level=0), cmap="YlGnBu", linewidths=1, square=True)
xticks_labels = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
plt.xticks(np.arange(12)+.5, labels=xticks_labels)
plt.show()
```
