```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0)
df.head()
```

```python
df.shape
```

```python
sns.lineplot(data=df)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title("Number of visitors to the museum")
plt.xlabel("sana")
plt.ylabel("odamlar soni")
plt.xticks(rotation=90)
sns.lineplot(data=df)
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.title("Muzeylarga tashriuflar soni")
plt.xlabel("sana")
plt.ylabel("odamlar soni")
plt.xticks(rotation=90)
plt.grid()
sns.lineplot(data=df['America Tropical Interpretive Center'])
plt.show()
```
